```python
import PyPDF2 
pdfFileObj = open(r'C:\Users\sunan\Downloads\CV\SUNANDA__RESUME.pdf', 'rb') 
pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 

# extracting text from page 
str = " "
for i in range(1):
    str += pdfReader.getPage(i).extractText()
with open(r"C:\Users\sunan\Downloads\CV\Output.txt", "w", encoding = 'utf-8') as f:
    f.write(str)

```
